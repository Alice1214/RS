## 中国主要城市房价、租金数据集

主要城市指以下19个城市：北京、上海、广州、深圳、成都、重庆、长沙、大连、合肥、杭州、济南、南京、青岛、沈阳、天津、武汉、厦门、烟台、郑州  

数据时间跨度为：2014.3 - 2019.3  

指标释义：
* 供给：市场供给的房产出售、出租数据；
* 关注：从城市房产网站获取的买房者或租房者的浏览轨迹分析出的关注热点；
* 价值：该月平均租金(元/月/m^2) * 12个月 * 30年；
* 售租比：房价/租金/12，单位为年；  

文件结构：
* Prices.xlsx：主要城市的房价走势数据。
* Rents.xlsx：主要城市的租金走势数据。
* /区房价：北上广深四大一线城市行政区划级别的房价走势数据。
* /区租金：北上广深四大一线城市行政区划级别的租金走势数据。

## 导入相关库

```python
# !/usr/bin/env python
# -*- coding=utf-8 -*-
import pandas as pd
import numpy as np
import time, datetime
from datetime import datetime, timedelta

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
#"inline"可直接在python console里面生成图像
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签

import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARMA, ARIMA
from itertools import product
import calendar
```

## 数据处理

```python
# 数据加载
df = pd.read_excel('./creprice_HousePriceData/Prices.xlsx') # 默认的就是第0个sheet,即北京
df = df[['时间(年-月)', '供给(元/㎡)']]
df.rename(columns = {"时间(年-月)": "Timestamp",'供给(元/㎡)':"Price"},inplace=True)

# 将时间作为df的索引
df.index = df.Timestamp

# 数据探索
print(df.head())
df.info()

# 按照月，季度，年来统计
df_month = df.resample('M').mean()
df_Q = df.resample('Q-DEC').mean()
df_year = df.resample('A-DEC').mean()

# 按照月，季度，年来显示房价的走势
fig = plt.figure(figsize=[15, 7])
plt.suptitle('房价', fontsize=20)
plt.subplot(221)
plt.plot(df_month.Price, '-', label='按月')
plt.legend()
plt.subplot(222)
plt.plot(df_Q.Price, '-', label='按季度')
plt.legend()
plt.subplot(223)
plt.plot(df_year.Price, '-', label='按年')
plt.legend()
plt.show()

# df所有行按时间索引升序重新排列，避免下面绘图时间轴是逆序的
df.sort_index(inplace=True)

#  返回三个部分 trend（趋势），seasonal（季节性）和residual (残留)
rs = sm.tsa.seasonal_decompose(df.Price, freq=12)
rs.plot()
plt.show()
```

## 使用ARIMA模型预测房价及结果可视化

```python
# 设置参数范围
ps = range(0, 3)
qs = range(0, 3)
ds = range(1, 2)
parameters = product(ps, ds, qs)
parameters_list = list(parameters)
# 寻找最优ARMA模型参数，即best_aic最小
results = []
best_aic = float("inf") # 正无穷
for param in parameters_list:
    try:
        #model = ARIMA(df_month.Price,order=(param[0], param[1], param[2])).fit()
        # SARIMAX 包含季节趋势因素的ARIMA模型
        model = sm.tsa.statespace.SARIMAX(df_month.Price,
                                order=(param[0], param[1], param[2]),
                                #seasonal_order=(4, 1, 2, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False).fit()

    except ValueError:
        print('参数错误:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
# 输出最优模型
print('最优模型: ', best_model.summary())


# 设置future_month，需要预测的时间date_list
df_month2 = df_month[['Price']]
future_month = 3
last_month = pd.to_datetime(df_month2.index[len(df_month2)-1])
#print(last_month)
date_list = []
for i in range(future_month):
    # 计算下个月有多少天
    year = last_month.year
    month = last_month.month
    if month == 12:
        month = 1
        year = year+1
    else:
        month = month + 1
    next_month_days = calendar.monthrange(year, month)[1]
    #print(next_month_days)
    last_month = last_month + timedelta(days=next_month_days)
    date_list.append(last_month)
print('date_list=', date_list)

# # 设置需要预测的时间date_list
# df_month2 = df_month[['Price']]
# date_list2 = [datetime.strptime('2019-04-01', '%Y-%m-%d'),datetime.strptime('2019-05-01', '%Y-%m-%d'),datetime.strptime('2019-06-01', '%Y-%m-%d')]

# 添加未来要预测的3年
future = pd.DataFrame(index=date_list, columns= df_month.columns)
df_month2 = pd.concat([df_month2, future])
df_month2['forecast'] = best_model.predict(start=0, end=len(df_month2))
# 第一个元素不正确，设置为NaN
df_month2['forecast'][0] = np.NaN
print(df_month2)


# 房价预测结果可视化
plt.figure(figsize=(30,7))
df_month2.Price.plot(label='实际房价')
df_month2.forecast.plot(color='r', ls='--', label='预测房价')
plt.legend()
plt.title('北京房价预测（月）')
plt.xlabel('时间')
plt.ylabel('房价')
plt.show()
```
